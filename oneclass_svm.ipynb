```python
import pandas as pd
import numpy as np
from sklearn.metrics import classification_report,accuracy_score,f1_score,
from sklearn.model_selection import train_test_split,GridSearchCV,cross_val_score
```

```python
#데이터 읽기
train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/머신러닝/project/Train.csv')
test = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/머신러닝/project/Test.csv')
```

```python
train=train.drop('Index',axis=1)
```

```python
#0은 negative, 1은 positive
for i in range(0,5000):
  if train.loc[i,'target']=='negative':
    train.loc[i,'target']= 1
  else:
    train.loc[i,'target']=-1
```

```python
neg = train[train['target']==1]
pos = train[train['target']==-1]
```

```python
X_train, X_test = train_test_split(neg.loc[:,'X0':'X999'],test_size=0.1,random_state=30)
```

```python
X_outliers = pos.loc[:,'X0':'X999']
X_outliers
```

```python
print("X_train_shape : ",X_train.shape)
print("X_test : ",X_test.shape)
print("X_outliers : ",X_outliers.shape)
```

```python
from sklearn.svm import OneClassSVM
clf = OneClassSVM(gamma=0.0001, kernel='linear', max_iter=1000, nu=0.5)

clf.fit(X_train)

y_pred_train = clf.predict(X_train)

y_pred_test = clf.predict(X_test)

y_pred_outliers = clf.predict(X_outliers)

n_error_train = y_pred_train[y_pred_train == -1].size

n_error_test = y_pred_test[y_pred_test == -1].size

n_error_outliers = y_pred_outliers[y_pred_outliers == -1].size
```

```python
print("train에서 -1로 예측한 수 : ", n_error_train)
print("error %: ",(n_error_train/4107))

print("test에서 -1로 예측한 수 : ", n_error_test)
print("error %: ",(n_error_test/457))

print("y_pred_outliers : ",y_pred_outliers)

print("-1을 -1로 분석한 수 : ", n_error_outliers)
print("acc %: ",(n_error_outliers/436))
```
